```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib
%matplotlib inline
import warnings
warnings.simplefilter('ignore')
```

```python
X = 2 * np.random.rand(100,1)
y = 4 + 3 * X + np.random.rand(100,1)
plt.plot(X,y,'r.')
plt.axis([0,2,0,15])
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

### Analytical Solution

#### Calculate Theta

```python
X_b = np.c_[np.ones((100,1)), X]
theta = np.linalg.inv(X_b.T.dot(X_b)).dot(X_b.T).dot(y)
theta
```

#### Predict new samples

```python
X_new = np.array([[0],[2]])
X_new_b = np.c_[np.ones((2,1)), X_new]
y_pred = X_new_b.dot(theta)
y_pred
```

```python
plt.plot(X,y,'r.')
plt.plot(X_new, y_pred, 'b--')
plt.axis([0,2,0,15])
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```
